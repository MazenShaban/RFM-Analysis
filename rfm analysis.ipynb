```python
df['Date'].max() # Find the maximum date from the column, 'Date'
```

```python
present = dt.datetime(2017, 1, 1) # set a specific date as 'present'
```

```python
rfm = df.groupby('Customer_ID').agg({
    # Get the no. of days between present and the max date for each group.
    'Date': lambda date: (present - date.max()).days,
    # Get the number of distinct values in the 'Transaction_ID' column for each group.
    'Transaction_ID': lambda num: num.nunique(),
    # Get the sum of all values in 'Sales_Amount' column for each group.
    'Sales_Amount': lambda totalPrice: totalPrice.sum()
})
```

```python
rfm
```

```python
rfm.columns = ['Recency', 'Frequency', 'Monetary'] # Rename the columns of 'rfm'
rfm
```

```python
# Destribute the values of each to their suitable quartile (1,2,3,4) '1' for the best and '4' for the worst
rfm['R_quartile'] = pd.qcut(rfm['Recency'], 4, labels = [1,2,3,4])
# rank(method='first') is for ensuring that each value has a unique rank, thereby, avoiding the issue of duplicted bin edges
rfm['F_quartile'] = pd.qcut(rfm['Frequency'].rank(method='first'), 4, labels = [4,3,2,1])
rfm['M_quartile'] = pd.qcut(rfm['Monetary'], 4, labels = [4,3,2,1])
```

```python
rfm
```

```python
# Concatenate all three quartiles in a single column.
rfm['rfm_Score'] = rfm['R_quartile'].astype(str) + rfm['F_quartile'].astype(str) + rfm['M_quartile'].astype(str)
rfm
```

```python
# Sort the RFM score to the highest priority customers first
rfm.sort_values('rfm_Score').head(10)
```
